```python
import csv
import random
import time
import sys
import os

def getInfo():
    
    """
    Gets user input for first and last name, and their ID. The ID is the checked if it is valid through another function.
    If the ID is not valid, the user has 3 attempts to input a valid ID. 
    If not valid after 3 attempts, the program will exit.
    """
    
    fname=input("First Name: ").title()  #Capitalizes first letter if not already
    lname=input("Last Name: ").title()   #Capitalizes first letter if not already

    attempts = 0
    while attempts < 3:
        userID=input("Enter ID: ")
        #passes userID into validateID function to test whether valid or not
        if validateID(userID) == True:
            return fname, lname, userID
        else:
            print("Inavlid ID. Try again.")
            attempts += 1
    if attempts == 3:
        print("Too many attempts. Exiting Program")
        sys.exit()

def validateID(userID):
    """
    Takes the userID as an argument, then goes through a series of if statements to check if ID is valid.
    """
    if len(userID) != 6:              #checks if ID is 6 characters long
        return False
    if userID[0] != "A":              #checks if first character is A
        return False
    if "0" in userID[1:]:             #checks if 
        return False
    if not userID[1:].isdigit():      #checks if all of the characters after A are digits
        return False
    return True                       #if none of the if statements are True then ID is valid

def quizLength():
    """
    This function gets the users selection of the quiz length of either 10 or 20 questions.
    If the user selects 10 questions, each question is 1 point.
    If the user selects 20 questions, each question is 0.5 points.
    """
    
    while True:
        questionChoice = int(input("Choose the number of questions for your quiz (10 or 20): "))

        if questionChoice == 10:
            length = 10
            weight = 1
            return length, weight
        elif questionChoice == 20:
            length = 20
            weight = 0.5
            return length, weight
        else:
            print("Invalid number of questions")

def getQuestions():
    """
    Opens the csv file containing all of the questions, answer choices, and correct answer.
    The csv file is then read into a list of lists called testBank for the program to use.
    """
    
    testBank = []
    with open("testBank.csv", newline="") as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            testBank.append(row)
    return testBank

def displayQuestions(question, i):
    """
    This function is called from the quiz function, and is used to display a random question each time is is called
    
    """
    
    #displays randomly selected question
    print("\nQuestion", i, ": ", question[0])
    print("A: ", question[1])
    print("B: ", question[2]) 
    print("C: ", question[3])
    return question

def quiz(uniqueQuestions, weight):
    """
    This function is used to execute the quiz. It first creates a list to store all of the questions that are used 
    and a count of the correct answers to calculate the quiz score. The time is also started within this function
    for the duration of the quiz. The function takes the questions from the main function and the wight of the questions
    as arguments.
    """
    
    questionList = []
    correctAnswers = 0
    startTime = time.time()   #starts timer
 
    for i, question in enumerate(uniqueQuestions, start=1): 
        if time.time() - startTime > 600:                    #if timer goes over 10 minutes, quiz terminates
            print("Time is up! Terminating quiz.")
            break
        displayQuestions(question, i)
        while True:
            answer = input("Answer: ").upper()                        #gets answer input
            if answer != "A" and answer != "B" and answer != "C":     #validates answer choice
                print("Invalid answer. Try again.")
            else:
                break

        questionList.append((question, answer))        #adds question to questionList to display at end      

        if answer == question[4]:                      #count of correct answers
            correctAnswers += 1       
    
    score = (correctAnswers / len(uniqueQuestions)) * 100     #calculates score

    #ends timer then caclulates total time
    endTime = time.time()                        
    totalTime = (endTime-startTime) / 60

    return questionList, score, totalTime

def quizFinish(fname, lname, userID, totalTime, questionList, score, weight):
    """
    Once the quiz goes through all of the questions, this function is called to save all of the information into a text
    file. It takes multiple arguments that are all saved into the text file. 
    """
    
    #creates file and names it with userID, first and last name
    fileName = userID + "_" + fname + "_" + lname + ".txt"   

    #writes user's info and quiz info to the text file
    with open(fileName, "w") as file:                                
        file.write("User ID: " + userID + "\n") 
        file.write("Name: " + fname + " " + lname + "\n")
        file.write("Elapsed time: " + str(round(totalTime, 2)) + " minutes\n")
        file.write("Score: " + str(score) + "%\n\n" )

        #writes each question, correct answer, and user's answer to the text file
        for question, answer in questionList:
            file.write("Question: " + question[0] + "\n")
            file.write("Correct Answer: " + question[4] + "\n")
            file.write("Your Answer: " + answer + "\n\n")

def main():
    while True:   

        #gets user info from function
        fname, lname, userID = getInfo()
    
        #gets user choice of 10 or 20 questions from function
        numOfQuestions, weight = quizLength()
    
        #function called to read questions from the csv file and save them in a list called testBank
        testBank = getQuestions()
    
        #randomly selects questions from the text bank and based on the number of questions chosen
        uniqueQuestions = random.sample(testBank, numOfQuestions)
    
        #calls the quiz function to execut quiz using the unique questions and weight of each question
        #returns the list of questions used, score, and amount of time taken
        questionList, score, totalTime = quiz(uniqueQuestions, weight)
       
    
        print("\nQuiz complete.\n")
    
        quizFinish(fname, lname, userID, totalTime, questionList, score, weight)
    
        #asks user to either quit quiz or start a new quiz
        quitOrStop = input("Enter Q to exit quiz or S to start a new quiz: ")
        if quitOrStop.upper() == "Q":
            print("Exiting quiz. Goodbye!")
            break
        else:
            print("\n" * 50)
    
        

if __name__=="__main__":
    main()
```
